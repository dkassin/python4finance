# Complex Selection
## data selection is accomplished by formulation of conditions on column values, and potentially combining multiple such conditions logically

```python
data = np.random.standard_normal((10,2))
df = pd.DataFrame(data, columns=['x','y'])
df.info()
df.head()
df.tail()

# Sample logical choices

df['x'] > 0.5

# This uses boolean indexing(creates a series of booleans), then uses that to filter and only return rows that are true
print(df[df['x'] > 0])

# 
print(df.query('x > 0'))
```

# concatenation, Joining, and Merging 

```python
df1 = pd.DataFrame(['100','200','300','400'], index=['a','b','c','d'],columns=['A',])
df1
df2 = pd.DataFrame(['200','150','50'], index=['f','b','d'],columns=['B',])
df2

pd.concat((df1,df2), sort=False)

df1.join(df2)
df2.join(df1)

c = pd.Series([250,150,50], index=['b','d', 'c'])
df1['C'] = c 
df2['C'] = c

pd.merge(df1, df2)
pd.merge(df1, df2, on='C')
pd.merge(df1, df2, how='outer')
```
